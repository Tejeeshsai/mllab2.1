```python
import pandas as pd
import numpy as np
file_path = "Lab Session Data.xlsx"  
sheet_name = "Purchase data"        

# Read the Excel sheet
data = pd.read_excel(file_path, sheet_name=sheet_name)

# Extract inputs and outputs
features = data[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']].dropna().values
targets = data[['Payment (Rs)']].dropna().values

# Matrix properties
num_features = features.shape[1]
num_samples = features.shape[0]
rank_matrix = np.linalg.matrix_rank(features)
features_pinv = np.linalg.pinv(features)
price_estimates = features_pinv @ targets

# Output results
print("Input Space Dimension:", num_features)
print("Total Observations:", num_samples)
print("Rank of Input Matrix:", rank_matrix)
print("Estimated Prices per Unit (Candy, Mango, Milk):", price_estimates.flatten())
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
excel_path = r"Lab Session Data.xlsx" 
sheet_name = "Purchase data"
# Read data from the specified sheet
customer_data = pd.read_excel(excel_path, sheet_name=sheet_name)
customer_data.columns = customer_data.columns.str.strip()
# Create a binary label: 1 = High Spender, 0 = Low Spender (based on payment threshold)
customer_data['SpendingClass'] = customer_data['Payment (Rs)'].apply(lambda val: 1 if val > 200 else 0)
input_matrix = customer_data[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']].values
output_labels = customer_data['SpendingClass'].values
# Initialize and train the logistic regression model
spend_model = LogisticRegression()
spend_model.fit(input_matrix, output_labels)
predicted_output = spend_model.predict(input_matrix)
print("Classification Results:")
print(classification_report(output_labels, predicted_output, target_names=["Low Spender", "High Spender"], zero_division=0))
conf_matrix = confusion_matrix(output_labels, predicted_output)
sns.heatmap(conf_matrix, annot=True, cmap='coolwarm', fmt='d',
            xticklabels=["Predicted: Low", "Predicted: High"],
            yticklabels=["Actual: Low", "Actual: High"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Purchase Classification - Confusion Matrix")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import statistics
import matplotlib.pyplot as plt
excel_path = r"Lab Session Data.xlsx" 
sheet_name = "IRCTC Stock Price"
# Read and clean data
stock_df = pd.read_excel(excel_path, sheet_name=sheet_name)
stock_df.columns = stock_df.columns.str.strip()  # Remove extra spaces in column headers
stock_df['ChangePercent'] = stock_df['Chg%'].astype(str).str.replace('%', '').astype(float)
avg_stock_price = statistics.mean(stock_df['Price'])
price_var = statistics.variance(stock_df['Price'])
wednesday_data = stock_df[stock_df['Day'] == 'Wed']
avg_wed_price = statistics.mean(wednesday_data['Price'])
april_data = stock_df[stock_df['Month'] == 'Apr']
avg_april_price = statistics.mean(april_data['Price']) if not april_data.empty else None
days_with_loss = (stock_df['ChangePercent'] < 0).sum()
loss_chance = days_with_loss / len(stock_df)
positive_wed = (wednesday_data['ChangePercent'] > 0).sum()
wed_profit_chance = positive_wed / len(wednesday_data)
plt.figure(figsize=(8, 5))
plt.scatter(stock_df['Day'], stock_df['ChangePercent'], color='orange', edgecolor='black')
plt.title("Daily % Change in IRCTC Stock")
plt.xlabel("Weekday")
plt.ylabel("Change (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
print("Average Price of IRCTC:", round(avg_stock_price, 2))
print("Variance in Price:", round(price_var, 2))
print("Average Price on Wednesdays:", round(avg_wed_price, 2))
print("Average Price in April:", round(avg_april_price, 2) if avg_april_price else "No April data available")
print("Probability of Loss:", round(loss_chance, 3))
print("Probability of Profit on Wednesday:", round(wed_profit_chance, 3))
print("Conditional Probability (Profit | Wednesday):", round(wed_profit_chance, 3))
```

```python
import pandas as pd
import numpy as np
excel_path = r"Lab Session Data.xlsx"
sheet_name = "thyroid0387_UCI"

# Read and prepare data
thyroid_data = pd.read_excel(excel_path, sheet_name=sheet_name)
thyroid_data.columns = thyroid_data.columns.str.strip()
thyroid_data.replace('?', np.nan, inplace=True)
print("\nATTRIBUTE TYPES:")
binary_features = [col for col in thyroid_data.columns if set(thyroid_data[col].dropna().unique()) <= {'f', 't'}]
nominal_features = ['sex', 'referral source']
target_feature = 'Condition'
numeric_features = ['Record ID', 'age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI', 'TBG']
# Print categories
for col in numeric_features:
    print(f"{col} - Numeric")
for col in binary_features:
    print(f"{col} - Binary")
for col in nominal_features:
    print(f"{col} - Nominal")
print(f"{target_feature} - Target")

print("\nMISSING VALUES:")
for column in thyroid_data.columns:
    missing_count = thyroid_data[column].isna().sum()
    if missing_count > 0:
        print(f"{column}: {missing_count}")
valid_numeric_cols = [col for col in numeric_features if col in thyroid_data.columns]
for col in valid_numeric_cols:
    thyroid_data[col] = pd.to_numeric(thyroid_data[col], errors='coerce')
print("\nNUMERIC STATISTICS:")
for col in valid_numeric_cols:
    col_data = thyroid_data[col].dropna()
    if col_data.empty:
        continue
    min_val = col_data.min()
    max_val = col_data.max()
    mean_val = round(col_data.mean(), 2)
    std_dev = round(col_data.std(), 2)
    outlier_count = col_data[(col_data < mean_val - 3*std_dev) | (col_data > mean_val + 3*std_dev)].count()
    print(f"{col}  Min: {min_val}, Max: {max_val}, Mean: {mean_val}, Std: {std_dev}, Outliers: {outlier_count}")
```

```python
import random
def create_values(count):
    return [random.randint(1, 100) for _ in range(count)]
def compute_avg(values):
    return sum(values) / len(values)
def compute_middle(values):
    values.sort()
    length = len(values)
    return (values[length // 2] if length % 2 else (values[length // 2 - 1] + values[length // 2]) / 2)
def compute_common(values):
    freq_map = {}
    for val in values:
        freq_map[val] = freq_map.get(val, 0) + 1
    return max(freq_map, key=freq_map.get)
total = int(input("Enter total numbers to generate: "))
data = create_values(total)
print("Generated List:", data)
print("Average (Mean):", compute_avg(data))
print("Median:", compute_middle(data))
print("Most Frequent (Mode):", compute_common(data))
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_excel("Lab Session Data.xlsx", sheet_name="thyroid0387_UCI")
df = df.replace('?', np.nan)
df = df.replace({'f': 0, 't': 1})

numeric_df = df.select_dtypes(include=[np.number])
numeric_df = numeric_df.dropna()

v1 = numeric_df.iloc[0].values.reshape(1, -1)
v2 = numeric_df.iloc[1].values.reshape(1, -1)

similarity = cosine_similarity(v1, v2)[0][0]
print("Cosine Similarity:", round(similarity, 4))
```

```python
import pandas as pd
import numpy as np
from scipy.stats import mode
df = pd.read_excel("/mnt/data/Lab Session Data.xlsx", sheet_name="thyroid0387_UCI")
df.replace('?', np.nan, inplace=True)
print("\nImputing missing values...\n")
for column in df.columns:
    if df[column].isnull().sum() > 0:
        try:
            df[column] = pd.to_numeric(df[column], errors='coerce')
            if df[column].dtype in [np.float64, np.int64]:
                q1 = df[column].quantile(0.25)
                q3 = df[column].quantile(0.75)
                iqr = q3 - q1
                is_outlier = (df[column] < q1 - 1.5 * iqr) | (df[column] > q3 + 1.5 * iqr)

                if is_outlier.any():
                    median_val = df[column].median()
                    df[column].fillna(median_val, inplace=True)
                    print(f"{column} - Filled with Median")
                else:
                    mean_val = df[column].mean()
                    df[column].fillna(mean_val, inplace=True)
                    print(f"{column} - Filled with Mean")
            else:
                mode_val = mode(df[column].dropna())[0][0]
                df[column].fillna(mode_val, inplace=True)
                print(f"{column} - Filled with Mode")
        except:
            mode_val = mode(df[column].dropna())[0][0]
            df[column].fillna(mode_val, inplace=True)
            print(f"{column} - Filled with Mode")

print("\nImputation Done.")
```
